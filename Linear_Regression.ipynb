```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('Canada_PCI_data.csv')
df
```

```python
plt.xlabel('Year')
plt.ylabel('PCI (US$)')
plt.scatter(df.year,df[["PCI (US$)"]])
```

```python
new_df=df.drop('PCI (US$)',axis='columns')
new_df
```

```python
PCI=df[["PCI (US$)"]]
PCI
```

```python
reg=linear_model.LinearRegression()
reg.fit(new_df,PCI)
```

```python
reg.predict([[2020]])
```

```python
plt.xlabel('Year')
plt.ylabel('PCI (US$)')
plt.scatter(df.year,df[["PCI (US$)"]])
plt.plot(df.year,reg.predict(df[['year']]),color='red')
```
